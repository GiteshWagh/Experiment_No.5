```python
# Basics Of Set
# It is also a datatype. That store multiple items in single variables. Denoted by {}. Set is unordered.
ExampleSet = {"Python","Java","Ruby","R","C#","C++"} # Example

print(ExampleSet)
print(len(ExampleSet)) # Length of set
print(type(ExampleSet)) # Output : <class 'set'>

# Note : Values True & 1 is and False & 0 are considered as same in sets.
TestSet = {True,1,False,0}
print(TestSet)

ExampleSet2 = {1,"String",True,[1,2,3,4,5],3.14} # Sets supports all type of datatypes.
print(ExampleSet2)
```

```python
ExampleSet = {"Python","Java","Ruby","R","C#","C++"} # Example
ExampleSet2 = {1,"String",True,(1,2,3,4,5),3.14}
ExampleSet.add("Javascript") # Adding an item
print(ExampleSet)
ExampleSet.remove("Java") # Removing an item
print(ExampleSet)

# Set Operations
print("Union Of Sets : ", ExampleSet | ExampleSet2)
print("Intersection Of Sets : ", ExampleSet & ExampleSet2)
print()
```

```python
# Problem 5 :
# Given the set of all students and a set of present students, find the absent student .
# Concept used : Set Difference(-)

# Solution :
# Set of all students in class
AllStudents = {"Student1","Student2","Student3","Student4","Student5","Student6","Student7"}

# Set of all present students
PresentStudents = {"Student1","Student2","Student4"}

# Absent students are those who are not present in the set of present students.
AbsentStudents = AllStudents - PresentStudents
print("These students are absent : ",AbsentStudents) # List of absent students
```

```python
# Basics Of Dictionary
laptop = {
    "brand" : "Dell",
    "model" : "G15",
    "year" : 2022
}

print(laptop)

# Length of the dictionary
print(len(laptop))

# Type of variable
print(type(laptop))
# Output : <class 'dict'>

# Changing the values
laptop["year"] = 2023
print(laptop)

# Retrival of values
print("Brand : ",laptop['brand'])

# Adding a new item with kay
laptop["Price"] = 55000
print(laptop)

# Removing elements from list
laptop.pop("Price") # Price & 55000 removed from dictionary
print(laptop)
```

```python
# Problem 4 :
# A dictionary store student names and marks. Write a code to find the topper in the list of students.

# Soution :
Student = {
    "Student1" : "97",
    "Student2" : "88",
    "Student3" : "90",
    "Student4" : "80"
}

MaxMarks = max(Student.values())
topper = max(Student, key=Student.get)

print("Topper is : ",topper)
print("Marks : ",MaxMarks)
```
